```python
import requests
from bs4 import BeautifulSoup
import os

os.makedirs("mental_health_docs", exist_ok=True)

URLS = [
    "https://www.who.int/news-room/fact-sheets/detail/mental-health-strengthening-our-response",
    "https://www.who.int/news-room/fact-sheets/detail/depression",
    "https://www.who.int/news-room/fact-sheets/detail/anxiety-disorders"
]

def scrape_and_save(url, filename):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    paragraphs = soup.find_all("p")
    text = "\n".join([p.get_text() for p in paragraphs])

    with open(f"mental_health_docs/{filename}.txt", "w", encoding="utf-8") as f:
        f.write(text)

for i, url in enumerate(URLS):
    scrape_and_save(url, f"article_{i+1}")

print("Scraping Complete. Files saved in mental_health_docs/")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

documents = []
doc_dir = "mental_health_docs/"

for filename in os.listdir(doc_dir):
    with open(os.path.join(doc_dir, filename), "r", encoding="utf-8") as f:
        text = f.read()
        documents.append(text)

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_text("\n".join(documents))

with open("mental_health_docs/processed_texts.txt", "w", encoding="utf-8") as f:
    f.write("\n\n".join(chunks))

print(f"Processed {len(chunks)} text chunks for retrieval.")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

with open("mental_health_docs/processed_texts.txt", "r", encoding="utf-8") as f:
    chunks = f.readlines()

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vector_db = FAISS.from_texts(chunks, embedding_model)
vector_db.save_local("mental_health_vector_db")

print("Vector database created and saved.")
```

```python
import google.generativeai as genai
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
import faiss
import time
import re
from sklearn.metrics.pairwise import cosine_similarity
from langdetect import detect


genai.configure(api_key="AIzaSyBKJoz0rbdnWkG9sKnFj32x5f5oZhA2I-o")

embedding_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vector_db = FAISS.load_local("mental_health_vector_db", embedding_model, allow_dangerous_deserialization=True)

placeholder_doc = [Document(page_content="This is an empty chat history entry.", metadata={"type": "chat"})]
chat_db = FAISS.from_documents(placeholder_doc, embedding_model)

def is_allowed_query(user_input):
    critical_patterns = [r"suicide", r"harm (myself|me)", r"end my life", r"kill myself", r"severe depression"]
    technical_patterns = [r"\b(machine learning|AI|programming|technology|code)\b"]

    if any(re.search(pattern, user_input, re.IGNORECASE) for pattern in critical_patterns):
        return False, "I'm really sorry you're feeling this way. Please consider reaching out to a mental health professional or a trusted person. You are not alone."

    if any(re.search(pattern, user_input, re.IGNORECASE) for pattern in technical_patterns):
        return False, "I'm here to support mental health conversations. For technical queries, I recommend referring to relevant resources."

    return True, ""

def query_gemini(prompt, user_language):
    model = genai.GenerativeModel("gemini-1.5-flash")

    lang_prompt = f"Respond in {user_language}. " + prompt

    response = model.generate_content(lang_prompt)

    return response.text if hasattr(response, "text") else "I'm really sorry you're feeling this way. I'm here to listen and support you."

def extract_relevant_info(user_input, retrieved_docs):
    input_embedding = embedding_model.embed_query(user_input)
    doc_embeddings = [embedding_model.embed_query(doc.page_content) for doc in retrieved_docs]

    scores = cosine_similarity([input_embedding], doc_embeddings)[0]
    relevant_docs = [retrieved_docs[i].page_content for i in range(len(scores)) if scores[i] > 0.5]
    return " ".join(relevant_docs) if relevant_docs else user_input

def save_to_chat_db(user_input, bot_response):
    retrieved_docs = vector_db.similarity_search(user_input, k=3)
    refined_input = extract_relevant_info(user_input, retrieved_docs)

    if refined_input.strip():
        combined_text = f"User: {refined_input}\nBot: {bot_response}"
        doc = Document(page_content=combined_text, metadata={"type": "chat"})
        chat_db.add_documents([doc])

        chat_docs = chat_db.similarity_search("", k=200)
        if len(chat_docs) > 100:
            chat_db.delete([chat_docs[i].id for i in range(len(chat_docs) - 100)])

def simulate_typing_effect(text):
    import random
    for char in text:
        print(char, end="", flush=True)
        if char in ".!?":
            time.sleep(random.uniform(0.05, 0.1))
        else:
            time.sleep(random.uniform(0.01, 0.03))
    print("\n")

def rag_chatbot():
    print("\nMental Health Chatbot (type 'exit' to quit)\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("Bot: You're not alone. Take care and be kind to yourself.")
            break

        try:
            user_language = detect(user_input)
        except:
            user_language = "English"

        allowed, message = is_allowed_query(user_input)
        if not allowed:
            print(f"Bot: {message}\n")
            continue
        retrieved_docs = vector_db.similarity_search(user_input, k=3)
        chat_history = chat_db.similarity_search(user_input, k=2)

        retrieved_text = "\n\n".join([doc.page_content for doc in retrieved_docs])
        chat_context = "\n\n".join([doc.page_content for doc in chat_history])

        system_prompt = f"Here is the previous conversation context:\n{chat_context}\n\n" if chat_context else ""
        context = f"{retrieved_text}\n\n{chat_context}" if retrieved_text else chat_context

        prompt = (
            f"{system_prompt}I want you to respond in a kind, supportive, and understanding way only if the question is relevant to mental health, otherwise tell that the bot is only designed as a mental health assistant."
            f" Use the following information to provide helpful and empathetic responses, the answer should also be looking like how a profesional therapist would talk about these situations:\n{context}\n\nQ: {user_input}\nA:"
        )

        bot_response = query_gemini(prompt, user_language)

        print("Bot: ", end="")
        simulate_typing_effect(bot_response)

        save_to_chat_db(user_input, bot_response)

rag_chatbot()
```
